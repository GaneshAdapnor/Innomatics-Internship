```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        return "Recharge amount must be at least ₹50."
    elif amount not in valid_plans:
        return f"Invalid plan! Available plans are {valid_plans}"
    else:
        return "Recharge Successful!"

# Retry using while loop
while True:
    try:
        amount = int(input("Enter recharge amount: "))
        result = validate_recharge(amount)
        print(result)
        
        if result == "Recharge Successful!":
            break
        else:
            print("Please try again.\n")
            
    except ValueError:
        print("Invalid input! Enter numeric value.\n")

```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for item in products:
        if products[item] < 15:
            print(item, "→ Reorder Alert")
        else:
            print(item, "→ Stock OK")

inventory = {
    "Rice": 20,
    "Sugar": 8,
    "Oil": 12,
    "Soap": 25
}

check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def student_result(marks):
    total = 0
    
    for mark in marks:
        total += mark
        
    average = total / len(marks)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# User Input
marks_input = input("Enter marks separated by space: ").split()
marks = []

for m in marks_input:
    marks.append(float(m))

result = student_result(marks)
print("Result:", result)
```

```python
# 4. Cab Fare Estimator with Retry Option

# Function to calculate fare
def calculate_fare(distance, peak_hour):
    base_fare = 50 
    per_km = 12 

    # Calculate fare without peak hrs
    fare = base_fare + (per_km * distance)

    # 25% extra for peak hour
    if peak_hour.lower() == 'yes':
        fare += fare * 0.25

    return fare

# loop for retry
while True:
    distance = float(input("Enter distance traveled (in km): "))
    peak_hour = input("Is it peak hour? (Yes/No): ")

    total_fare = calculate_fare(distance, peak_hour)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate fare again? (Yes/No): ")
    if retry.lower() != 'yes':
        print("Thank you for using the Cab Fare Calculator!")
        break
```

```python
# Problem 5: Attendance Eligibility Checker

def check_attendance(attendance):
    total_days = len(attendance)
    present_days = 0
    
    for day in attendance:
        if day.upper() == "P":
            present_days += 1
    
    percentage = (present_days / total_days) * 100
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance_input = input("Enter attendance record (P/A): ")
result = check_attendance(attendance_input)

print("Attendance Status:", result)
```

```python
# Problem 6: Password Strength Checker

def password_strength(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
            
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

password = input("Enter password: ")
result = password_strength(password)

print(result)
```
