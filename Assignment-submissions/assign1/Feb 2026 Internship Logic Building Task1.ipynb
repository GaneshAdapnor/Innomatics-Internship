```python
correct_username = "admin"
correct_password = "1234"

entered_username = input("Enter username: ")
entered_password = input("Enter password: ")

if(entered_username == correct_username and entered_password == correct_password):
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0

for mark in marks:
    if(mark >= 50):
        pass_count += 1
    else:
        fail_count += 1

print("Total Students: ", len(marks))
print("Number of Passed Students: ",pass_count)
print("Number of Failed Students: ",fail_count)
```

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []

for name in names:
    name = name.strip() #For removing extra spaces
    name = name.lower() #For converting to lowercase
    cleaned_names.append(name) #Adding to cleaned names

print("Original Names:",names)
print("Cleaned Names:",cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
    message_length = len(message) #Storing length of message
    print("Message:",message)
    print("Message Length:",message_length)

    if(message_length > 10):
        print("Flagged")
    else:
        print("Not Flagged")
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0

for log in logs:
    if(log == "ERROR"):
        error_count += 1

print("System logs:",logs)
print("Total ERROR messages:",error_count)
```
