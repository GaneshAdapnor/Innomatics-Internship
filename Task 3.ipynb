```python
# Dictionary containing employee names and their scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85,
}

# Find the highest score among employees
highest_score = max(employees.values())

# Get list of employees who scored the highest marks
top_performers = [name for name, score in employees.items() if score == highest_score]

# Print result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

```python
# Import Counter from collections module
from collections import Counter

# Take user input and convert to lowercase
text = input("Enter a sentence: ").lower()

# Split sentence into words
words = text.split()

# Count frequency of each word
word_count = Counter(words)

# Filter words that appear more than once
repeated_words = {word: count for word, count in word_count.items() if count > 1}

# Print repeated words
print("Repeated Words:", repeated_words)
```

```python
# List of sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Find even readings along with their index position
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]

# Print valid readings
print("Valid Sensor Readings (Hour, Value):", valid_readings)
```

```python
# Import regex module
import re

# List of email addresses
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store domain count
domain_count = {}

# Regex pattern to extract domain name
pattern = r'@([\w\.]+)'

# Loop through emails
for email in emails:
    match = re.search(pattern, email)
    if match:
        domain = match.group(1)
        # Increase domain count
        domain_count[domain] = domain_count.get(domain, 0) + 1

# Total number of emails
total_emails = len(emails)

# Calculate and print percentage for each domain
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Import numpy for calculating mean
import numpy as np

# List of daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
average_sales = np.mean(sales)

# Print days where sales are greater than average
for day, amount in enumerate(sales):
    if amount > average_sales:
        print(f"Day {day + 1}: {amount}")
```

```python
# Import Counter
from collections import Counter

# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count occurrences of each user ID
counts = Counter(user_ids)

# Print only duplicate user IDs
for user, count in counts.items():
    if count > 1:
        print(f"{user} → {count}")
```
